## SciPy: Overview

[SciPy](https://scipy.org/) is an ecosystem of Python packages built for scientific computing. We'll focus on 3 packages:

* [NumPy](http://www.numpy.org/): fast array computations
* [Matplotlib](https://matplotlib.org/): 2D plotting (... and 3D)
* [SciPy](https://scipy.org/): library of scientific functions and stuff

Check out the [SciPy 2018](https://www.youtube.com/watch?v=y7zGnKzaKIw&list=PLYx7XA2nY5Gd-tNhm79CNMe_qvi35PgUR) conference videos.

Let's do a quick example. Import the `pandas` package under the name `pd` (standard convention).

```python
import pandas as pd
```

We're going to make a plot and so we use [Jupyter magic](https://ipython.readthedocs.io/en/stable/interactive/magics.html#magic-matplotlib) tell our Jupyter notebook to display plots in the notebook.

```python
%matplotlib inline
```

I have a few datasets on my webpage. We can give a URL to the `pandas.read_csv` function to download and import the data file. See more about `pandas.read_csv` in the [documentation](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_csv.html).

```python
df = pd.read_csv('https://www.math.ubc.ca/~pwalls/data/van_crime.csv')
```

We've saved the pandas DataFrame with the variable `df` and let's verify the type.

```python
type(df)
```

A DataFrame is like a spreadsheet. The columns are NumPy arrays of values. We can see the first 5 lines with the `head` method.

```python
df.head()
```

This is a dataset from the [Vancouver Police Department](https://data.vancouver.ca/datacatalogue/crime-data.htm). Each row is a reported crime. There are some crimes where the location is not included for privacy reasons and the coordaintes are set to 0.

```python
df[df['X'] == 0].head()
```

Let's plot the locations as a scatter plot. Note that `pandas` is using `matplotlib` behind the scenes.

```python
df[df['X'] > 0].plot(x='X',y='Y',kind='scatter',s=2,alpha=0.5,lw=0,figsize=(10,8))
```

We've set some keyword parameters to display the plot clearly. The parameter `s` is size, `alpha` is transparancy, `lw` is line wide of the border of the circle markers and `figsize` is the figure size.

## NumPy Arrays

A [Python package](https://docs.python.org/3/tutorial/modules.html#packages) is a directory of files containing Python code. We use the `import` statement to import a package.

Let's import NumPy with the name `np` (standard convention). We acces NumPy objects and functions by the dot syntax.

```python
import numpy as np
```

The most important new datatype provided by the NumPy package is the `numpy.ndarray`. We can create NumPy arrays in several ways. The easiest is the function `numpy.array` which converts any sequence into a NumPy array.

```python
x = np.array([1,2,3])
```

```python
type(x)
```

```python
x
```

```python
y = np.array(range(0,20))
```

```python
y
```

```python
type(y)
```

There are [several functions for creating arrays](https://docs.scipy.org/doc/numpy/user/quickstart.html#array-creation). Let's use `numpy.linspace` to create an array of 100 evenly spaced values from 0 to 5.

```python
x = np.linspace(0,5,100)
```

```python
type(x)
```

```python
x.size
```

Use array operations to create an array of $y$ values for the function $y = e^{-x^2}$.

```python
y = np.exp(-x**2)
```

```python
y.size
```

We can view the first 5 values:

```python
y[0:5]
```

Let's plot the arrays `x` versus `y` using matplotlib.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```
